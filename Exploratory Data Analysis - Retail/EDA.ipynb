## **State wise analysis of Profit, Discount, Sale**

```python
df['State'].value_counts().head(10)
```

```python
data_state= df.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
data_state.head(10)
```

```python
#State wise Profit analysis
data_state1=data_state.sort_values('Profit')
```

```python
data_state1[['Profit']].plot(kind = 'bar', figsize = (20,8))
plt.title('State wise Profit Analysis', fontsize = 20)
plt.ylabel('Profit per Sate')
plt.xlabel('States')
plt.show()
```

```python
data_state['Sales'].plot(kind='pie',figsize = (20,20),autopct='%1.1f%%',startangle=90,shadow=True)
plt.title('State wise analysis of Sale',fontsize=20)
```


1.Highest amount of sales= Wyoming(11.8%)

2.Lowest amount of sales= South Dakota(0.8%)

State wise Discount analysis

```python
data_state1['Discount'].plot(kind='bar',figsize=(18,5),color='orange')
plt.title('State wise analysis of Discount', fontsize=20)
```
